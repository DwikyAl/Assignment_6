```python
#Merubah dataframe ke array numpy
datasets = datasets.to_numpy() 
```

```python
print(datasets)
```

```python
#Membagi datasets => 400 baris data untuk training dan 100 baris data untuk testing
dataTraining = np.concatenate((datasets[0:120, :], datasets[150:269, :]), axis=0)
dataTesting = np.concatenate((datasets[120:150, :], datasets[269:299, :]), axis=0)
```

```python
print(dataTraining)
len(dataTraining)
```

```python
print(dataTesting)
len(dataTesting)
```

```python
#Memecah dataset ke input label
inputTraining = dataTraining[:,0:12]
inputTesting = dataTesting[:,0:12]
labelTraining = dataTraining[:,12]
labelTesting = dataTesting[:,12]
print(labelTraining)
len(labelTraining)
```

```python
#Mendefinisikan decision tree classifier
model = tree.DecisionTreeClassifier()
```

```python
#Mentraining model
model =  model.fit(inputTraining, labelTraining)
```

```python
#Memprediksi input data testing
hasilPrediksi = model.predict(inputTesting)
print("Label Sebenarnya : ", labelTesting)
print("Hasil Prediksi : ", hasilPrediksi)
```

```python
#Menghitung akurasi
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
print("Prediksi Benar : ", prediksiBenar, " Data")
print("Prediksi Salah : ", prediksiSalah, " Data")
print("Akurasi : ", prediksiBenar/(prediksiBenar+prediksiSalah) * 100, "%")
```
